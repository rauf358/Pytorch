### **Transfer Learning: A Smart Way to Train Models Faster**  

**What is Transfer Learning?**  
Transfer learning allows us to **reuse** the knowledge (weights) of a model trained on one problem and apply it to a different but related problem.  

For example:  
- A model trained on **millions of images** (e.g., ImageNet) has already learned to detect basic shapes, edges, and textures.  
- We can **use this knowledge** and fine-tune the model to classify our **own images** (like pizza, steak, and sushi) instead of training from scratch.  

---

### **Why Use Transfer Learning?**  
‚úÖ **Saves time & resources** ‚Äì No need to train from scratch on large datasets.  
‚úÖ **Better performance with less data** ‚Äì Since the model has already learned useful patterns.  
‚úÖ **Works well for small datasets** ‚Äì Even if you have limited training images, transfer learning helps.  

---

### **How It Works in PyTorch?**  
1Ô∏è‚É£ **Load a Pretrained Model** (e.g., ResNet, VGG, EfficientNet).  
2Ô∏è‚É£ **Freeze some layers** (to keep previously learned features).  
3Ô∏è‚É£ **Modify the final layers** to fit your dataset.  
4Ô∏è‚É£ **Train on your custom dataset** to fine-tune the model.  

This way, you build a **high-performance model quickly** without needing millions of images. üöÄ

# Creating a transform for torchvision.models (auto creation)


### **Auto-creation of Transforms in torchvision**  
When using a **pretrained model** from `torchvision.models`, it is crucial that your **custom dataset** is preprocessed in the **same way** as the dataset used to train the model originally (e.g., ImageNet).  

### **Why is this important?**  
Pretrained models learn **features** from images with specific preprocessing (e.g., normalization, resizing). If your data doesn‚Äôt follow the same process, it could lead to **poor predictions**.

### **Automatic Transform Creation (torchvision v0.13+)**  
Instead of manually defining transforms, `torchvision.models` provides an **automatic transform** feature.  

#### **How does it work?**
1Ô∏è‚É£ Choose a pretrained model and **get its default weights**  
   ```python
   weights = torchvision.models.EfficientNet_B0_Weights.DEFAULT
   ```
   - `EfficientNet_B0_Weights`: Defines the **pretrained weights** for EfficientNet-B0.  
   - `DEFAULT`: Chooses the **best available** weights.

2Ô∏è‚É£ Get the **corresponding transforms** automatically  
   ```python
   preprocess = weights.transforms()
   ```
   - `preprocess` ensures that the **same preprocessing** used during training (ImageNet) is applied to new images.  

### **Advantages of Auto-Transforms**
‚úÖ No need to manually write transforms  
‚úÖ Ensures **consistent preprocessing** for better performance  
‚úÖ Supports multiple pretrained models in `torchvision.models`  

This feature **simplifies transfer learning** by handling preprocessing automatically! üöÄ

````python
import os
import shutil
import random
from pathlib import Path

# Define source directories
source_dirs = {
    "glaucoma": Path("data/dataset/glaucoma"),
    "normal": Path("data/dataset/normal")
}

# Define destination directories
dest_dirs = {
    "train": {
        "glaucoma": Path("data/dataset/train/glaucoma"),
        "normal": Path("data/dataset/train/normal"),
    },
    "test": {
        "glaucoma": Path("data/dataset/test/glaucoma"),
        "normal": Path("data/dataset/test/normal"),
    }
}

# Ensure destination directories exist
for category in dest_dirs.values():
    for path in category.values():
        os.makedirs(path, exist_ok=True)

# Function to split and move images
def split_and_move_images(category):
    image_paths = list(source_dirs[category].glob("*.jpg"))  # Get all image files

    if len(image_paths) == 0:  # Check if images exist
        print(f"No images found in {source_dirs[category]}. Skipping...")
        return

    random.shuffle(image_paths)  # Shuffle images

    # Split 80% for training, 20% for testing
    split_idx = int(0.8 * len(image_paths))
    train_images = image_paths[:split_idx]
    test_images = image_paths[split_idx:]

    # Move images to respective folders
    for img in train_images:
        shutil.move(str(img), str(dest_dirs["train"][category] / img.name))

    for img in test_images:
        shutil.move(str(img), str(dest_dirs["test"][category] / img.name))

    print(f"Successfully moved {len(train_images)} train & {len(test_images)} test images for '{category}'.")

# Apply the function to both categories
split_and_move_images("glaucoma")
split_and_move_images("normal")

print("\n Dataset successfully split into 80% train and 20% test.")
````

### **Breaking Down EfficientNet_B0**
Your **EfficientNet_B0** model consists of three key parts:

#### **1Ô∏è‚É£ Feature Extractor (`features`)**
- A stack of **convolutional layers**, **batch normalization**, and **activation functions**.
- These layers help extract **low-level** (edges, textures) and **high-level** (shapes, objects) features from images.
- This part is **often kept frozen** (weights unchanged) when applying transfer learning.

#### **2Ô∏è‚É£ Adaptive Pooling (`avgpool`)**
- Converts the extracted features into a **fixed-size feature vector**.
- This is essential because different images may have different dimensions, but a **consistent feature vector size** is needed for classification.

#### **3Ô∏è‚É£ Classifier (`classifier`)**
- A fully connected (`Linear`) layer that **maps the feature vector to output classes**.
- The default model has `out_features=1000` (for ImageNet classes).
- When applying transfer learning, this layer is **replaced** with a new one that matches the number of classes in your custom dataset.

# Freezing the base model and changing the output layer to suit our needs


### **How to Freeze Layers in PyTorch (EfficientNet Example)**  

Since EfficientNet was pretrained on **ImageNet (1000 classes)** but we only need **3 classes (pizza, steak, sushi)**, we **freeze** its feature extractor while modifying the classifier layer.  

#### **Steps to Freeze Layers**  
1. Set `requires_grad = False` for all layers in the **features** section.  
2. Replace the final classifier layer (`nn.Linear`) to match our **3-class problem** (changing `out_features=1000` to `out_features=3`).  
3. Ensure that only the new classifier layers have `requires_grad = True` so that they get updated during training.

---

### **PyTorch Implementation**  
```python
import torch
import torchvision
import torch.nn as nn

# Load EfficientNet-B0 Pretrained Model
model = torchvision.models.efficientnet_b0(weights=torchvision.models.EfficientNet_B0_Weights.DEFAULT)

# Freeze all feature extractor layers
for param in model.features.parameters():
    param.requires_grad = False

# Modify classifier layer (1000 ‚Üí 3 classes)
model.classifier = nn.Sequential(
    nn.Dropout(p=0.2, inplace=True),  # Same as original
    nn.Linear(in_features=1280, out_features=3)  # Changed output classes
)

# Print model summary to confirm changes
print(model)
```

---

### **Why Do This?**
- The **features** layers (pretrained on ImageNet) stay **frozen**, preserving learned patterns.  
- Only the **classifier** learns new patterns specific to **pizza, steak, sushi**.  
- This makes training **faster** and prevents overfitting on small datasets.

### **Understanding Dropout in Neural Networks**  

Dropout is a **regularization technique** that helps prevent overfitting by randomly removing (or "dropping") some connections between neurons during training.  

#### **How Dropout Works:**
1. **Each forward pass:** A random subset of neurons is **turned off** (set to zero).  
2. **Each backward pass:** Only the active neurons update their weights.  
3. **During inference:** Dropout is **disabled**, and all neurons are used, but their activations are scaled to compensate for the dropped connections during training.  

---

### **Example: Dropout in EfficientNet**
EfficientNet uses **Dropout(p=0.2)** in the classifier:  
```python
nn.Dropout(p=0.2, inplace=True)
```
- `p=0.2` ‚Üí **20% of neurons** in this layer are randomly dropped during training.  
- `inplace=True` ‚Üí The operation modifies the data directly to save memory.  

---

### **Why Use Dropout?**
‚úÖ **Prevents overfitting** ‚Üí Forces the model to learn **robust, general features** instead of memorizing patterns.  
‚úÖ **Adds randomness** ‚Üí Encourages the network to find multiple independent representations.  
‚úÖ **Improves generalization** ‚Üí Helps the model perform well on unseen data.

### **Key Takeaways from Freezing Layers in Transfer Learning**  

When modifying a **pretrained model** for your task (like EfficientNet for classifying pizza, steak, and sushi), you generally **freeze the feature extractor** and **fine-tune only the classifier**. Here‚Äôs why:  

---

### **1. Trainable vs. Non-Trainable Parameters**
- The **feature extractor (base layers)** has `requires_grad=False` ‚Üí It **won‚Äôt be updated** during training.  
- The **classifier (final layer)** has `requires_grad=True` ‚Üí It **will be trained** for your specific task.  

**üìå Benefit:** We keep the powerful pre-learned features from ImageNet and only train the last layer for our new dataset.

---

### **2. Output Shape Changes**
- Before: **[32, 1000]** (ImageNet has 1000 classes)  
- Now: **[32, 3]** (We only need 3 classes: pizza, steak, sushi)  

**üìå Benefit:** The model now aligns with our classification task.

---

### **3. Reduced Trainable Parameters = Faster Training**
- Before: **5.28 million trainable parameters**  
- Now: **Only 3,843 trainable parameters**  
  *(Most of the model remains frozen except for the classifier.)*  

**üìå Benefit:**  
‚úÖ **Less computation required** ‚Üí Faster training  
‚úÖ **Prevents overfitting** ‚Üí The model already knows useful visual features  
‚úÖ **Works well with smaller datasets**

#Train function

````python
def train_step(
    model:nn.Module,
    dataloader:torch.utils.data.DataLoader,
    loss_fn:torch.nn.Module,
    optimizer:torch.optim.Optimizer,
    accuracy_fn:accuracy_fn,
    device:torch.device
):
  model.train()
  train_loss,train_acc=0,0
  for batch,(X,y) in enumerate(dataloader):
    X,y=X.to(device),y.to(device)
    y_pred_logits=model(X)
    y_pred=torch.round(torch.sigmoid(y_pred_logits))
    loss=loss_fn(y_pred_logits,y.type(torch.float).unsqueeze(1))
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    train_loss+=loss.item()
    train_acc += (y_pred == y.unsqueeze(1)).sum().item()/len(y)
  train_loss=train_loss/len(dataloader)
  train_acc=train_acc/len(dataloader)
  return train_loss,train_acc
````

````python
def test_step(
    model:nn.Module,
    dataloader:torch.utils.data.DataLoader,
    loss_fn:torch.nn.Module,
    accuracy_fn:accuracy_fn,
    device:torch.device):
  model.eval()
  test_loss,test_acc=0,0
  with torch.inference_mode():
    for batch,(X,y) in enumerate(dataloader):
      X,y=X.to(device),y.to(device)
      test_pred_logits=model(X)
      test_pred=torch.round(torch.sigmoid(test_pred_logits))
      loss=loss_fn(test_pred_logits,y.type(torch.float).unsqueeze(1))
      test_loss+=loss.item()
      test_acc += ((test_pred == y.unsqueeze(1)).sum().item()/len(y))
    test_loss=test_loss/len(dataloader)
    test_acc=test_acc/len(dataloader)
  return test_loss,test_acc
````

#main loop

````python

epochs=5
results={
    "train_loss":[],
    "train_acc":[],
    "test_loss":[],
    "test_acc":[]
}
train_loss_values=[]
train_acc_values=[]
test_loss_values=[]
test_acc_values=[]
from tqdm.auto import tqdm
from timeit import default_timer as timer
start_time=timer()
for epoch in tqdm(range(epochs)):
  print(f"Epoch:{epoch}\n-------")
  train_loss_values,train_acc_values=train_step(
      model=model,
      dataloader=train_dataloader,
      loss_fn=loss_fn,
      optimizer=optimizer,
      accuracy_fn=accuracy_fn,
      device=device)

  test_loss_values,test_acc_values=test_step(
      model,
      test_dataloader,
      loss_fn,
      accuracy_fn=accuracy_fn,
      device=device)
  results["train_loss"].append(train_loss_values)
  results["train_acc"].append(train_acc_values)
  results["test_loss"].append(test_loss_values)
  results["test_acc"].append(test_acc_values)
  print(f"train_loss:{train_loss_values: .4f} train_acc:{train_acc_values: .4f}")
  print(f"test_loss:{test_loss_values: .4f} test_acc:{test_acc_values: .4f}")
end_time=timer()
print(f"total time taken:{end_time-start_time}")
````

Here‚Äôs the **conceptual flow** of how the `pred_and_plot_image()` function should work:

1. **Open and Prepare the Image**  
   Use a library like PIL to open the image, and make sure it's in RGB format. Then apply the **same transform** used during training (e.g., resizing, normalization).

2. **Format the Image for the Model**  
   Add a batch dimension to the image (so it matches the `[batch_size, channels, height, width]` shape), convert it to a float tensor, and send it to the **same device** as the model.

3. **Switch the Model to Evaluation Mode**  
   This ensures layers like dropout or batchnorm behave correctly. Also use inference mode to disable gradient tracking, which speeds things up.

4. **Make the Prediction**  
   Pass the image through the model. Since this is binary classification, the model will output a single logit. Apply a **sigmoid function** to get the prediction probability, and then **round it** to get the final class label (0 or 1).

5. **Interpret the Result**  
   Use the class label to find the corresponding class name from your `class_names` list. Also, grab the prediction probability for context.

6. **Visualize**  
   Show the image using matplotlib or another tool, and annotate it with the predicted class and its probability.

````python
from typing import List,Tuple
import matplotlib.pyplot as plt
from PIL import Image

def pred_and_plot_Image(
    model:nn.Module,
    image_path:str,
    class_names:List[str],
    image_size:Tuple[int,int]=(224,224),
    device:torch.device=device):
  image=Image.open(image_path)
  model.to(device)
  model.eval()
  with torch.inference_mode():
    transformed_image=auto_transform(image)
    transformed_image=transformed_image.unsqueeze(0)
    transformed_image=transformed_image.to(device)
    target_image_pred=model(transformed_image)
    target_image_pred_probs=torch.sigmoid(target_image_pred)
    target_image_pred_label=torch.round(target_image_pred_probs)

    pred_class_index = round(target_image_pred_label.item()) #added this line to get the index of the predicted label
    #print(f"target_image_pred_label:{target_image_pred_label}, pred_class_index:{pred_class_index} ")
    plt.figure()
    plt.imshow(image)
    plt.title(f"Pred: {class_names[pred_class_index]} | Prob: {target_image_pred_probs.max():.3f}")
    plt.axis(False);
````
