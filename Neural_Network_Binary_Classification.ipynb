### ‚úÖ **PyTorch Neural Network Classification Overview**

---

## üéØ **What is a Classification Problem?**
A **classification problem** is when the **model predicts discrete labels or categories** rather than continuous values (like in regression).

---

## üåü **Three Main Types of Classification Problems:**

| **Type of Classification** | What is it?                                                    | Example |
|----------------|------------------------------------------------------------------------|-----------------------------|
| **Binary Classification** | Classifying between **two possible outcomes** (0 or 1, yes or no). | Predict whether a person has heart disease (yes/no). |
| **Multi-Class Classification** | Classifying between **more than two classes**, but only **one output** is selected. | Identifying an image as either a **cat, dog, or bird**. |
| **Multi-Label Classification** | An input can **belong to multiple categories at once**. | Predicting topics for a Wikipedia article (e.g., **Science, Math, and Technology**). |

---

## ‚úÖ **How PyTorch Handles Classification:**
1Ô∏è‚É£ **Input data (features)** ‚ûú Pass through the model.  
2Ô∏è‚É£ **Model output (logits)** ‚ûú Raw scores before applying softmax or sigmoid.  
3Ô∏è‚É£ **Activation function** ‚ûú Converts logits into probabilities.  
   - `sigmoid()` for **binary classification**.  
   - `softmax()` for **multi-class classification**.  
4Ô∏è‚É£ **Loss function** ‚ûú Measures how far predictions are from the actual labels.  
   - `nn.BCELoss()` for **binary**.  
   - `nn.CrossEntropyLoss()` for **multi-class**.  
5Ô∏è‚É£ **Optimizer** ‚ûú Updates the model's weights using gradient descent.  

---

## ‚úÖ **A Quick Example of Multi-Class Classification:**
Imagine you're predicting **handwritten digits (0-9)** from an image.

- **Input**: Image pixels.  
- **Output**: Probabilities for each digit (0 to 9).  
- **Loss function**: `nn.CrossEntropyLoss()`.  
- **Optimizer**: `torch.optim.SGD()` or `torch.optim.Adam()`.

  \[
  [\text{None}, 3, 224, 224]
  \]
  or  
  \[
  [32, 3, 224, 224]
  \]
  where **32 is the batch size** (number of images processed at once).

---

### üìå **Output Shape:**
- After passing the input through the **machine learning model**, you get **prediction probabilities** for each class.  
  For example, in the image, the model predicts:  
  \[
  [0.97, 0.00, 0.03]
  \]
  indicating **97% confidence for class 1 (burger)**, **0% for class 2 (steak)**, and **3% for class 3 (pizza)**.

- The **output shape** is:  
  \[
  [3]
  \]
  because there are **3 possible classes**.

---

|-----------------|----------------|-------------------------|
| Input Tensor  | `[batch_size, 3, 224, 224]` | Represents multiple images |

### **üîπ Understanding Input and Output Shapes in Deep Learning**  
One of the most common mistakes in deep learning is **shape mismatches**. If tensor dimensions don‚Äôt align, your model will break.  

To **avoid shape errors**, you should always ask:  
üëâ *‚ÄúWhat is the shape of my inputs and what is the expected shape of my outputs?‚Äù*  

---

### **üîπ How to Think About Shapes in Deep Learning?**  
1Ô∏è‚É£ **Input Shape** ‚Üí The shape of the data fed into the model.  
2Ô∏è‚É£ **Hidden Layers** ‚Üí Intermediate transformations between input and output.  
3Ô∏è‚É£ **Output Shape** ‚Üí The shape of the final prediction.  

#### **Example: Image Classification (CNNs)**
- **Input Shape:** `[batch_size, channels, height, width]` ‚Üí `[32, 3, 224, 224]`
- **Output Shape:** `[batch_size, num_classes]` ‚Üí `[32, 3]` (if classifying between food, dog, or person)  

#### **Example: Tabular Data Classification (MLPs)**
- **Input Shape:** `[batch_size, num_features]` ‚Üí `[32, 5]` (if using 5 features like age, height, etc.)  
- **Output Shape:**  
  - `[batch_size, 1]` for **binary classification**  
  - `[batch_size, num_classes]` for **multiclass classification**  

---

### **üîπ Common Shape Errors & Fixes**
‚úÖ **Mismatch Between Layers** ‚Üí Ensure correct dimensions between input, hidden, and output layers.  
‚úÖ **Incorrect Batch Size Handling** ‚Üí Use `reshape()` or `unsqueeze()` if needed.  
‚úÖ **Inconsistent Feature Shapes** ‚Üí Check `.shape` after every transformation.

### **üîπ Why Do We Define a `forward()` Method in PyTorch?**  

In PyTorch, **defining the `forward()` method** is required because it **tells the model how data should flow through the network** during a **forward pass** (making predictions).  

---
### **üîπ Key Reasons We Define `forward()`**
1Ô∏è‚É£ **Controls the Computation**  
   - Defines how input data is processed through layers.
   - You decide the sequence of operations (e.g., Linear layers, activation functions, etc.).

2Ô∏è‚É£ **Allows Customization**  
   - You can modify layer connections, apply activation functions, or implement advanced architectures.

3Ô∏è‚É£ **Required for Training**  
   - PyTorch needs `forward()` to **compute the loss** and update weights in **backpropagation**.

---
### **üîπ Example: Why Can't We Just Use `__init__()`?**
‚úÖ **`__init__()` only defines the layers** but doesn‚Äôt specify how data flows through them.  
‚úÖ **`forward()` defines how layers interact** and what transformations occur.  

#### **Example: Without `forward()`**
```python
class CircleModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer_1 = nn.Linear(2, 5)
        self.layer_2 = nn.Linear(5, 1)
```
‚ùå This model **only defines the layers** but does not tell PyTorch **how to use them** for predictions.

---
### **üîπ Example: Defining `forward()` Properly**
```python
class CircleModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer_1 = nn.Linear(2, 5)
        self.layer_2 = nn.Linear(5, 1)

    def forward(self, x):
        return self.layer_2(self.layer_1(x))  # Defines forward pass
```
‚úÖ Now, PyTorch **knows** how to process the input data (`x`) through `layer_1` and `layer_2`.  

---
### **üîπ What Happens When You Call `model(x)`?**
When you run:
```python
model = CircleModel()
output = model(X_sample)  # Calls forward() internally
```
‚úÖ PyTorch **automatically calls `forward()`** to generate predictions.  
‚úÖ This is why you **never call `model.forward(x)` directly**‚ÄîPyTorch handles it internally.

---

|-------------|----------------|
| **Defines computation** | Tells PyTorch how to pass data through layers |

| **Required for training** | Needed for loss computation & backpropagation |

### **üîπ Difference Between `nn.Module` Subclassing and `nn.Sequential` in PyTorch**  

Both **subclassing `nn.Module`** and using **`nn.Sequential`** allow you to build neural networks, but they have key differences.

---

## **1Ô∏è‚É£ Using `nn.Sequential()`**
‚úÖ **When to Use?**  
- When you have a simple, linear stack of layers with **no complex logic**.  
- The layers **execute in the exact order** they are listed.

### **Example: Using `nn.Sequential`**
```python
model_0 = nn.Sequential(
    nn.Linear(in_features=2, out_features=5),
    nn.Linear(in_features=5, out_features=1)
).to(device)
```
üîπ **What Happens?**
- Input `x` passes **directly** from `Linear(2 ‚Üí 5)` ‚Üí `Linear(5 ‚Üí 1)`.  
- No need to explicitly define `forward()`, PyTorch **automatically** applies the layers in sequence.

---

## **2Ô∏è‚É£ Subclassing `nn.Module`**
‚úÖ **When to Use?**  
- When you need a **custom forward pass** (e.g., activation functions, branching, skip connections).  
- When the model is **not just a simple stack of layers**.

### **Example: Subclassing `nn.Module`**
```python
class CircleModelV0(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer_1 = nn.Linear(in_features=2, out_features=5)
        self.layer_2 = nn.Linear(in_features=5, out_features=1)

    def forward(self, x):
        x = self.layer_1(x)  # First layer
        x = torch.relu(x)    # Adding activation function
        x = self.layer_2(x)  # Second layer
        return x  # Output
```
üîπ **What Happens?**
- **Explicit control** over **how data flows** through the model.
- We **added ReLU activation**, which **can‚Äôt** be done with `nn.Sequential` (unless manually added).
- Can include **conditional logic** (e.g., if statements).

---

## **3Ô∏è‚É£ Key Differences**
| Feature | `nn.Sequential` | Subclassing `nn.Module` |
|---------|----------------|--------------------------|
| **Ease of Use** | ‚úÖ Simple to write | ‚ùå More code needed |
| **Custom `forward()` Logic** | ‚ùå Not possible | ‚úÖ Fully customizable |
| **Adding Activation Functions** | ‚ùå Must include them in `Sequential()` | ‚úÖ Can add in `forward()` |
| **Skip Connections (ResNets, etc.)** | ‚ùå Not possible | ‚úÖ Possible |
| **Multiple Inputs/Outputs** | ‚ùå Difficult | ‚úÖ Easy |

### **Understanding the Accuracy Function in PyTorch**
The function `accuracy_fn` calculates how many predictions are correct out of the total number of predictions. Here's a breakdown of how it works:

```python
def accuracy_fn(y_true, y_pred):
    correct = torch.eq(y_true, y_pred).sum().item()
    acc = correct / len(y_pred)
    return acc
```

---

### **Step-by-Step Explanation**
1. **`torch.eq(y_true, y_pred)`**  
   - Compares `y_true` (actual labels) and `y_pred` (model predictions) **element-wise**.
   - Returns a tensor of `True` (1) or `False` (0), where `True` means the prediction was correct.

2. **`.sum().item()`**  
   - `sum()` counts the number of `True` values (i.e., correct predictions).  
   - `.item()` converts the tensor to a Python integer.

3. **`acc = correct / len(y_pred)`**  
   - The number of correct predictions is divided by the total number of predictions to get the accuracy.

4. **Returns accuracy as a fraction (between 0 and 1).**  
   - Multiply by 100 if you want percentage accuracy.

---

### **‚ö†Ô∏è Issue with `accuracy_fn`**
This function assumes `y_pred` contains **exact class labels (0 or 1)**. However, in PyTorch, model outputs for **classification tasks** are often raw logits or probabilities.

To fix this, **apply thresholding for binary classification**:
```python
def accuracy_fn(y_true, y_pred):
    y_pred_labels = torch.round(torch.sigmoid(y_pred))  # Convert logits to 0 or 1
    correct = torch.eq(y_true, y_pred_labels).sum().item()
    acc = correct / len(y_pred)
    return acc
```

For **multi-class classification**, use `torch.argmax()`:
```python
def accuracy_fn(y_true, y_pred):
    y_pred_labels = torch.argmax(y_pred, dim=1)  # Get class with highest probability
    correct = torch.eq(y_true, y_pred_labels).sum().item()
    acc = correct / len(y_pred)
    return acc
```

---

### **‚úÖ Summary**
- Use `torch.eq()` to compare predictions and actual labels.
- Apply `torch.sigmoid()` and `torch.round()` for **binary classification**.
- Use `torch.argmax()` for **multi-class classification**.
- Multiply by `100` if you want accuracy in **percentage form**.

üöÄ **This ensures the accuracy function correctly evaluates classification models!**

### **What Are Logits in Deep Learning?**

In deep learning, **logits** refer to the raw, unnormalized outputs of a neural network before applying an activation function like **sigmoid** or **softmax**. They represent scores assigned by the model to different classes but are not directly interpretable as probabilities.

---

### **Why Are Logits Used?**
1. **Numerical Stability:**  
   - Using logits avoids issues with floating-point precision when computing loss functions like **cross-entropy**.
2. **Flexible Activation:**  
   - Logits allow us to apply different activation functions depending on the task (e.g., **sigmoid** for binary classification, **softmax** for multi-class classification).
3. **Direct Input to Loss Functions:**  
   - PyTorch's `torch.nn.BCEWithLogitsLoss()` and `torch.nn.CrossEntropyLoss()` expect logits because they internally apply the necessary activation functions.

---

### **Logits in Binary Classification**
For **binary classification**, the output layer has **one neuron** without an activation function:
```python
output = model(x)  # Logits (raw scores, can be any real number)
```
To convert logits into probabilities, we apply the **sigmoid function**:
```python
probabilities = torch.sigmoid(output)  # Converts logits into values between 0 and 1
```
Then, we classify based on a **threshold** (e.g., `0.5`):
```python
predictions = (probabilities > 0.5).float()  # Converts to 0 or 1
```

---

### **Logits in Multi-Class Classification**
For **multi-class classification**, the output layer has **multiple neurons (one per class)**, producing logits:
```python
output = model(x)  # Logits (e.g., 3 values for 3 classes)
```
To convert these logits into probabilities, we apply **softmax**:
```python
probabilities = torch.softmax(output, dim=1)  # Converts logits into probabilities (sum to 1)
```
Predicted class:
```python
predictions = torch.argmax(probabilities, dim=1)  # Picks the class with the highest probability
```

---

### **Example: Why Not Use Softmax in the Model Directly?**
PyTorch‚Äôs `torch.nn.CrossEntropyLoss()` expects **logits** as input because it applies `softmax()` internally for better numerical stability. If you apply softmax **before** passing logits to `CrossEntropyLoss`, the loss computation may become unstable.

‚úÖ **Correct:**
```python
loss_fn = torch.nn.CrossEntropyLoss()
loss = loss_fn(logits, y_true)  # Logits are used directly
```

‚ùå **Incorrect:**
```python
softmax_outputs = torch.softmax(logits, dim=1)
loss = loss_fn(softmax_outputs, y_true)  # This is incorrect, loss expects logits!
```

---

### **üîë Key Takeaways**
- **Logits** are raw scores from a neural network before activation.
- **Sigmoid** (for binary classification) and **softmax** (for multi-class classification) convert logits into probabilities.
- **Loss functions like `BCEWithLogitsLoss()` and `CrossEntropyLoss()` expect logits** for better numerical stability.
- When predicting, apply the appropriate activation function **outside** the loss function to interpret results.

üöÄ **Understanding logits helps prevent common deep learning mistakes!**

### **üîß Hyperparameters in Deep Learning**
Hyperparameters are **tunable settings** in a neural network that affect how the model learns. They **aren't learned from data**‚Äîinstead, you set them manually and adjust through experimentation.  

---

### **üîπ Common Hyperparameters**
| **Hyperparameter** | **What it Does** | **Typical Values** |
|--------------------|-----------------|--------------------|
| **Learning Rate (`lr`)** | Controls **how much** weights change per update. | `0.001` to `0.1` (lower for deep models). |
| **Batch Size** | Number of samples per training batch. | `16`, `32`, `64`, `128` (power of 2). |
| **Number of Epochs** | How many times the model sees the full dataset. | `10 - 1000` (depends on data size). |
| **Hidden Layers** | Number of layers between input and output. | `1 - 10+` (more layers for complex problems). |
| **Neurons per Layer** | Number of neurons per hidden layer. | `8 - 512+` (wider layers for more learning). |
| **Activation Function** | Introduces non-linearity to the network. | `ReLU`, `Sigmoid`, `Tanh`, `LeakyReLU`. |
| **Optimizer** | Determines **how** weights are updated. | `SGD`, `Adam`, `RMSprop`. |
| **Loss Function** | Measures model error. | `BCEWithLogitsLoss` (binary), `CrossEntropyLoss` (multi-class). |

---

### **üîπ Why Hyperparameters Matter**
1. **Too high a learning rate** ‚Üí Model jumps around and never converges.  
2. **Too low a learning rate** ‚Üí Model learns too slowly or gets stuck.  
3. **Too few epochs** ‚Üí Model underfits (doesn't learn enough).  
4. **Too many epochs** ‚Üí Model overfits (memorizes training data).  
5. **Too small a batch size** ‚Üí Training is slow and unstable.  
6. **Too large a batch size** ‚Üí Model may generalize poorly.  

---

### **üîπ Tuning Hyperparameters**
- **Manual Tuning** ‚Üí Adjust hyperparameters based on experience.  
- **Grid Search** ‚Üí Try different values systematically.  
- **Random Search** ‚Üí Try random combinations of values.  
- **Bayesian Optimization** ‚Üí Uses probability to find the best settings.

# 6. The missing piece: non-linearity
# We've seen our model can draw straight (linear) lines, thanks to its linear layers.

But how about we give it the capacity to draw non-straight (non-linear) lines?

 Why is Non-Linearity Important?
Without non-linearity ‚Üí The model can only learn simple relationships (straight lines).

With non-linearity ‚Üí The model can learn curved and more complex decision boundaries.

Stacking multiple layers without activation functions ‚Üí The entire network behaves like a single linear function (no real depth advantage!).



### **üîπ What is a Non-Linear Function?**
A **non-linear function** is a function **whose graph is not a straight line**. This means the relationship between the input and output **changes** in a complex way, rather than being proportional.

In mathematical terms, a function is **linear** if it satisfies both:
1. **Additivity:** \( f(x + y) = f(x) + f(y) \)
2. **Homogeneity:** \( f(a \cdot x) = a \cdot f(x) \)  

If a function **does not** satisfy both properties, it is **non-linear**.

---

### **üîπ Examples of Linear vs. Non-Linear Functions**
| **Type**      | **Function** | **Graph Shape** |
|--------------|-------------|----------------|
| **Linear**   | \( f(x) = 2x + 3 \) | Straight Line |
| **Non-Linear** | \( f(x) = x^2 \) | Parabola (Curved) |
| **Non-Linear** | \( f(x) = \sin(x) \) | Wave (Oscillating) |
| **Non-Linear** | \( f(x) = e^x \) | Exponential Curve |

---

### **üîπ Why Do We Need Non-Linearity in Neural Networks?**
A neural network with **only linear functions** (i.e., only `nn.Linear` layers) behaves like a **single linear transformation**, no matter how many layers you add. This means it can **only learn straight-line relationships**.

To learn **complex patterns**, we add **non-linear activation functions** like:
- **ReLU** ‚Üí \( f(x) = \max(0, x) \) (helps capture non-linearity)
- **Sigmoid** ‚Üí \( f(x) = \frac{1}{1 + e^{-x}} \) (useful for probabilities)
- **Tanh** ‚Üí \( f(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} \) (maps values between -1 and 1)

---

### **üîπ Real-World Example**
üîπ Imagine you want a neural network to classify images of **cats vs. dogs** üê±üê∂.  
If the model is **only linear**, it will try to separate cat and dog images with a straight line, which is **not enough** for real-world complexity.  

By adding **non-linearity**, the model can draw **complex decision boundaries** and learn subtle features like:
‚úÖ **Cat ears vs. Dog ears**  
‚úÖ **Whiskers vs. No Whiskers**  
‚úÖ **Different fur textures**  

---

### **üîπ Conclusion**
- **Linear functions** ‚Üí Can only model **simple, straight-line** relationships.
- **Non-linear functions** ‚Üí Can model **complex, curved** relationships.
- **Neural networks use non-linearity** (activation functions) to learn **better patterns** in data.

# building a model with non liearity

### **üöÄ Building a Model with Non-Linearity**
So far, our model has only used **linear transformations** (`nn.Linear`), which means it can **only** learn patterns that are **separable by straight lines**. But our **circle dataset** is **non-linear**, meaning it **can‚Äôt be separated by a straight line**.

### **üîπ Why Do We Need Non-Linearity?**
Linear layers alone **cannot** capture complex patterns. **Non-linearity** allows the model to learn **curved decision boundaries**, which is crucial for tasks like:
- Image recognition (detecting edges, textures, objects)
- Speech recognition (understanding voice variations)
- Classification of **non-linearly separable data** (like circles!)

---

### **üîπ Adding Non-Linearity with Activation Functions**
To introduce non-linearity, we place **activation functions** between linear layers. PyTorch provides several built-in options, with one of the most popular being **ReLU (Rectified Linear Unit)**.

#### **‚úÖ ReLU (Rectified Linear Unit)**
Formula:  
\[
f(x) = \max(0, x)
\]
- If **x is positive**, it stays the same.
- If **x is negative**, it becomes **0**.

#### **Why ReLU?**
‚úÖ It **removes negative values** (helps with sparsity).  
‚úÖ It is **computationally efficient** (faster training).  
‚úÖ It helps prevent the **vanishing gradient problem** (common in deep networks).

---

### **üîπ Implementing Non-Linearity in Our Model**
Let's modify our previous model to include **ReLU** after each linear layer:

```python
import torch.nn as nn

# Define a model with non-linearity
class CircleModelV1(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer_1 = nn.Linear(in_features=2, out_features=10)  # Increase hidden units
        self.layer_2 = nn.Linear(in_features=10, out_features=10)  # Another hidden layer
        self.layer_3 = nn.Linear(in_features=10, out_features=1)  # Output layer
        self.relu = nn.ReLU()  # Activation function

    def forward(self, x):
        x = self.relu(self.layer_1(x))  # Apply ReLU after layer 1
        x = self.relu(self.layer_2(x))  # Apply ReLU after layer 2
        x = self.layer_3(x)  # Output layer (no activation for now)
        return x

# Instantiate the model
model_1 = CircleModelV1()
```

‚úÖ Our model can now learn **curved** decision boundaries.  
‚úÖ More layers + **ReLU** = **Better learning capacity**.
