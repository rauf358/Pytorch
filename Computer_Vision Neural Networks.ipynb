# Q: What is Convolutional Neural Network ?

### **📌 What is a Convolutional Neural Network (CNN)?**  

A **Convolutional Neural Network (CNN)** is a type of deep learning model specifically designed for processing structured grid-like data, such as images. It is widely used in **computer vision** tasks like image classification, object detection, and facial recognition.

---

## **🔹 Why CNN Instead of a Regular Neural Network?**
A traditional neural network (**Fully Connected Network**) treats an image as a **1D vector**, losing spatial information.  
CNNs, however:
✔ Preserve **spatial relationships** in images.  
✔ Extract **features** automatically.  
✔ Reduce the number of **trainable parameters**, making them more efficient.  

---

## **🔹 CNN Architecture**
A CNN consists of **three main layers**:

### **1️⃣ Convolutional Layer (Feature Extraction)**
- Applies **filters (kernels)** to detect edges, textures, and patterns.
- Uses **convolution operations** to slide filters over the image.
- Produces a **feature map** highlighting important features.

📌 **Example:**  
🔲🔲🔲 → Edge detector filter → ▨▨▨ (feature map)

---

### **2️⃣ Pooling Layer (Dimensionality Reduction)**
- Reduces the size of the feature map while keeping important information.
- **Max pooling** is commonly used (takes the maximum value in a region).
- Helps in reducing **computation** and making the model more robust to **small changes**.

📌 **Example:**  
🔲🔲 → **Max Pooling (2×2)** → 🔲 (smaller feature map)

---

### **3️⃣ Fully Connected Layer (Classification)**
- After extracting features, the output is flattened into a 1D vector.
- Fully connected layers **map features to output classes** (e.g., cat vs. dog).
- Uses **softmax activation** to assign probabilities to classes.

📌 **Example:**  
🐱 90% 🐶 10% → Model predicts **"Cat"**.

---

## **🔹 How CNN Works Step by Step**
1. **Input Image** → A raw image is given as input (e.g., 28×28 grayscale or 3-channel RGB).
2. **Convolutional Layers** → Extracts different levels of **features** (edges, textures, shapes).
3. **Pooling Layers** → Reduces **size** while keeping essential information.
4. **Flattening** → Converts the feature map into a 1D vector.
5. **Fully Connected Layer** → Makes final predictions using softmax activation.
6. **Output Layer** → Classifies into categories (e.g., "dog" or "cat").

---

## **🔹 Why CNN is Powerful?**
✅ **Automatic Feature Extraction** → No need for manual feature engineering.  
✅ **Parameter Efficiency** → Reduces computation compared to fully connected networks.  
✅ **Translation Invariance** → Recognizes objects despite position changes.  

---

## **🔹 Applications of CNN**
- **Image Classification** (e.g., Face Recognition)  
- **Object Detection** (e.g., Self-Driving Cars)  
- **Medical Imaging** (e.g., Tumor Detection)  
- **Handwriting Recognition** (e.g., Digit Recognition)  

---

### **✅ Summary**
🔹 **CNNs are specialized for image data** and maintain spatial relationships.  
🔹 **Key layers**: Convolutional → Pooling → Fully Connected.  
🔹 **Used in** image recognition, object detection, and medical analysis.

## **📌 4. Using `torch.utils.data.Dataset` to Load Custom Data**
When working with your own dataset, you need to **create a custom dataset class** by subclassing `torch.utils.data.Dataset`.

### ✅ **Example: Creating a Custom Dataset Class**
```python
from torch.utils.data import Dataset
from PIL import Image
import os

class CustomImageDataset(Dataset):
    def __init__(self, image_dir, transform=None):
        self.image_dir = image_dir
        self.transform = transform
        self.image_files = os.listdir(image_dir)

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        img_path = os.path.join(self.image_dir, self.image_files[idx])
        image = Image.open(img_path).convert("RGB")
        if self.transform:
            image = self.transform(image)
        return image
```
🔹 This **custom dataset class**:
- Loads images from a directory.
- Applies **transformations** (if provided).
- Returns an image when indexed.

#Data loader and Mini batches

### **📌 Understanding DataLoader in PyTorch**
The **DataLoader** is a key component in PyTorch for **loading and processing datasets efficiently**.

---

### **🔹 Why Use a DataLoader?**
1. **Handles Large Datasets Efficiently**  
   - Instead of loading all data at once, it **breaks it into smaller batches**.  
   - Each batch is sent to the model one at a time, **saving memory and computation power**.

2. **Enables Mini-Batch Training**  
   - Instead of computing gradients on the whole dataset, mini-batches allow updates per batch.  
   - This **speeds up learning** and **improves generalization**.

3. **Shuffling for Better Training**  
   - Shuffling ensures that the model **doesn't memorize the order of data**, making training more robust.

4. **Parallel Processing**  
   - Using `num_workers`, the DataLoader can **load data in parallel**, reducing bottlenecks.

---

### **🔹 Creating a DataLoader**
In PyTorch, we can easily create a DataLoader from a dataset.

```python
from torch.utils.data import DataLoader

# Create DataLoaders for training and testing
BATCH_SIZE = 32  # Number of samples per batch

train_dataloader = DataLoader(
    dataset=train_data,  # Our training dataset
    batch_size=BATCH_SIZE,  # Load 32 samples per batch
    shuffle=True,  # Shuffle data each epoch
)

test_dataloader = DataLoader(
    dataset=test_data,
    batch_size=BATCH_SIZE,
    shuffle=False  # No need to shuffle test data
)
```

---

### **🔹 How Mini-Batches Work**
If our dataset has **60,000 images** and we set `batch_size=32`:
- Each batch contains **32 images**.
- **One epoch** (full dataset pass) will have **60,000 / 32 ≈ 1,875** batches.
- The model updates **1,875 times per epoch**, rather than once.

---

### **🔹 Choosing the Right Batch Size**
The batch size is a **hyperparameter** that you can experiment with.  
- **Small batch sizes (e.g., 16, 32):** More updates per epoch, better generalization.
- **Large batch sizes (e.g., 128, 256, 512):** Faster training but requires more memory.

✔ **32 is a good starting point!**  
✔ Try different values to see what works best for your model.

#Why Use nn.Flatten() in PyTorch?
 The nn.Flatten() layer is used to convert a multi-dimensional tensor (like an image) into a one-dimensional feature vector that can be processed by fully connected (nn.Linear) layers.


 ### **📌 Why Do We Get One Value Per Pixel in Our Output Vector?**
When we apply `nn.Flatten()` to an image, it **converts** the multi-dimensional pixel grid into a **1D vector**, where **each pixel corresponds to one value** in that vector.

---

### **📌 What Does This Mean?**
Each **pixel intensity** (grayscale or RGB value) is stored as a **single number** (usually between 0 and 1 if normalized).  

- In a **grayscale image** (1 channel), each pixel has **one intensity value** (e.g., 0 for black, 1 for white).
- In a **color image** (3 channels: RGB), each pixel has **three values** (one for Red, Green, and Blue).

🔹 Example:
- **Before Flattening:** A `28x28` grayscale image has **784 pixels**.
- **After Flattening:** It becomes a `1D vector` of size **784** (each number represents a pixel intensity).

---

### **📌 What Happens After Flattening?**
1. **Pixels Become Features**  
   - Each pixel is treated as an **individual feature** for the model to learn from.
   
2. **Connected to a Fully Connected Layer (`nn.Linear`)**  
   - This layer learns **patterns** between pixels.

3. **Pattern Recognition Begins**  
   - Instead of seeing pixels as a grid, the model finds **relationships** between pixel values to make predictions.

---

````python
from torch import nn
class FashionMNISTmodelV0(nn.Module):
  def __init__(self,input_shape:int,
               hidden_units:int,
               output_shape:int):
    super().__init__()
    self.layer_stack=nn.Sequential(
        nn.Flatten(),
        nn.Linear(in_features=input_shape,
                  out_features=hidden_units),
        nn.Linear(in_features=hidden_units,
                  out_features=output_shape)
    )

  def forward(self,x):
    return self.layer_stack(x)
````

````python
#import tqdm for progress bar
from tqdm.auto import tqdm
#set the seed and start the timer
torch.manual_seed(42)
train_time_start_on_cpu=timer()

#set the number of epochs (we ll keep this small for faster training time)
epochs=3
#create trianig and testing loop
for epoch in tqdm(range(epochs)):
  print(f"epoch:{epoch}\n------")
  #train
  train_loss=0# to calculate training loss per batch acumulatly add training loss
  #add a loop to loop trough the training batches
  for batch,(X,y) in enumerate(train_dataloader):
    model0.train()
    #forwardpass
    y_pred=model0(X)
    #calculate loss per batch
    loss=loss_fn(y_pred,y)
    train_loss+=loss#acumulate the training loss per batch
    #optimizer zero grad
    optimizer.zero_grad()
    #loss backward
    loss.backward()
    #optimizer step
    optimizer.step()
    #the optimizer will update the models perameters once per batch rather then once per epoch
    #print whats happeing?
    if batch %400==0:
      print(f'looked at {batch*len(X)}/{len(train_dataloader.dataset)} samples.')

  #devide total train loss by length of train dataloader
  train_loss /= len(train_dataloader)

  #testing loop
  test_loss,test_acc=0,0
  model0.eval()
  with torch.inference_mode():
    for X_test,y_test in test_dataloader:
      #forward pass
      test_pred=model0(X_test)

      #calculte loss acumulately
      test_loss+=loss_fn(test_pred,y_test)
      #calculate accuracy
      test_acc+=accuracy_fn(y_true=y_test,
                            y_pred=test_pred.argmax(dim=1))
  #calculate test loss avg per batch
    test_loss/=len(test_dataloader)
  #calculate test acc avg per batch
    test_acc/=len(test_dataloader)

  #print wahts hapnig
  print(f"\nTrain loss:{train_loss:0.4f}| Test loss:{test_loss:.4f}, test_acc{test_acc:.4f}")
  #calulate training time
  train_time_end_on_cpu=timer()
  total_train_time_model0=print_train_time(start=train_time_start_on_cpu,
                                           end=train_time_end_on_cpu,
                                           device=str(next(model0.parameters()).device))

````

### **Understanding the Training and Testing Loops in Mini-Batch Training**

Let's break it down step by step with an **example**.

---

### **Given Information**
- **Total Training Images**: **60,000**
- **Total Testing Images**: **10,000**
- **Batch Size**: **32**
- **Epochs**: **3**

---

### **How Many Batches?**
Since we divide the dataset into batches of **32 images per batch**, we calculate:

#### **Training Batches per Epoch**
\[
\frac{60000 \text{ images}}{32 \text{ images per batch}} = 1875 \text{ batches per epoch}
\]
So, in **one epoch**, the training loop runs **1875 times**.

#### **Testing Batches per Epoch**
\[
\frac{10000}{32} \approx 312 \text{ batches per epoch}
\]
So, in **one epoch**, the testing loop runs **312 times**.

---

### **Step-by-Step Breakdown of the Loops**

#### **Training Loop (Iterates Over Training Batches)**
```python
for epoch in range(epochs):  # Outer loop (runs 3 times)
    for batch, (X, y) in enumerate(train_dataloader):  # Inner loop (runs 1875 times per epoch)
```
- The **outer loop** runs **3 times** (for each epoch).
- The **inner loop** runs **1875 times per epoch**, loading one batch (32 images) at a time.
- Each iteration of the inner loop:
  - Loads a **new batch of 32 images**.
  - Passes them through the model.
  - Calculates **loss** and updates model weights.
- After **1875 batches**, one full epoch is complete.

**🔹 Example (1st Epoch, First Few Batches)**
| Batch Number | Images Loaded | Cumulative Images Processed |
|-------------|--------------|----------------------------|
| 1           | 32           | 32 |
| 2           | 32           | 64 |
| 3           | 32           | 96 |

## **What is a Convolutional Neural Network (CNN)?**
A **Convolutional Neural Network (CNN)** is a special type of **neural network** that is designed specifically for **image processing** and other visual tasks. It can recognize patterns, shapes, and objects in images, making it widely used in **image classification, object detection, and facial recognition**.

### **How does a CNN work?**
A CNN works by passing an image through multiple **layers**, each of which helps extract important features from the image. The structure follows this pattern:

**Input layer** → [**Convolutional layer** → **Activation function** → **Pooling layer**] → **Fully connected layer (Output layer)**

Each of the components plays a unique role in processing the image.

---

## **Components of a CNN**
Let’s go through each component in detail:

### **1. Input Layer**
- This is the **starting point** of the CNN.
- It takes in the raw image (e.g., a colored image with **Red, Green, and Blue (RGB) channels**).
- The input is represented as a **3D tensor** (height, width, and depth for RGB channels).

### **2. Convolutional Layers (Feature Extractors)**
- This is the **most important part of a CNN**.
- It applies **filters (kernels)** to the image to extract features such as **edges, textures, and patterns**.
- Each neuron in the convolutional layer performs a **dot product** between its filter and the previous layer's output.
- The result is called an **activation map**.

#### **Hyperparameters of Convolutional Layers**
1. **Kernel Size (Filter Size)**
   - Defines the size of the filter that moves over the image.
   - A **small kernel size** (e.g., 3x3) captures fine details.
   - A **large kernel size** (e.g., 5x5) captures more general patterns.

2. **Stride**
   - Defines how many pixels the filter moves at a time.
   - A **small stride** (e.g., 1) captures more detailed features.
   - A **large stride** (e.g., 2) reduces computation but might lose details.

3. **Padding**
   - Helps preserve spatial information by adding extra pixels around the image.
   - The most common type is **zero-padding**, where zeros are added around the edges.
   - Padding ensures the filter can cover all parts of the image without shrinking it too much.

---

### **3. Activation Function (ReLU)**
- After each convolution, an **activation function** is applied to introduce **non-linearity**.
- CNNs commonly use **ReLU (Rectified Linear Unit)** because:
  - It removes negative values (sets them to zero).
  - It helps CNNs train faster and improves accuracy.
- Formula:  
  \[
  f(x) = \max(0, x)
  \]
  - If \( x > 0 \), it remains the same.
  - If \( x \leq 0 \), it becomes 0.

---

### **4. Pooling Layer (Downsampling)**
- This layer **reduces the size of activation maps**, making computation more efficient.
- The most common pooling method is **Max-Pooling**, which:
  - Uses a small window (e.g., 2x2) and selects the **maximum value** in that window.
  - Reduces **dimensions** while keeping the most important features.
- Pooling helps in:
  - Reducing computation.
  - Preventing **overfitting**.
  - Making CNNs **invariant to small shifts** in the image.

---

### **5. Fully Connected Layer (Classification) Claasifier layer**
- After extracting features, CNNs use **fully connected layers** to classify images.
- The **Flatten Layer** converts the 3D feature maps into a **1D vector** for classification.
- A **Softmax function** is applied at the end to convert scores into **probabilities**.

#### **Softmax Function**
- Converts raw scores into probabilities.
- Ensures that the sum of all outputs equals **1**.
- Helps the model decide which class the image belongs to.

---

## **How CNNs Learn? (Training Process)**
1. **Forward Propagation**:  
   - The input image passes through all layers, and the model makes a prediction.
2. **Loss Calculation**:  
   - The difference between the predicted output and the correct label is measured using a **loss function**.
3. **Backpropagation**:  
   - The CNN **adjusts the weights** (kernels) using **gradient descent** to minimize errors.
4. **Repetition**:  
   - The process repeats over many images to make the CNN more accurate.

---

## **Choosing the Right Model**
Depending on the type of data, different models are used:

| **Problem Type**         | **Best Model Type**               |
|--------------------------|----------------------------------|
| **Structured Data** (Excel, tables) | Gradient Boosting, Random Forests (XGBoost) |
| **Unstructured Data** (Images, audio, text) | CNNs, Transformers |

For **image classification**, CNNs like **AlexNet, VGG, ResNet, and EfficientNet** are commonly used.

---

## **Summary**
A **CNN** is a powerful deep learning model for **image classification** and other vision-related tasks. It consists of:
1. **Convolutional Layers** to extract important features.
2. **Activation Functions (ReLU)** to introduce non-linearity.
3. **Pooling Layers** to reduce dimensions.
4. **Fully Connected Layers** to classify the features.
5. **Softmax Function** to output class probabilities.

### **🔹 Key Hyperparameters in CNNs**  

CNNs have three key hyperparameters that **control how convolution operations** extract features from an image:  
1️⃣ **Padding**  
2️⃣ **Kernel Size**  
3️⃣ **Stride**  

---

## **1️⃣ Padding: Preserving Borders**  
🔹 **What is it?**  
- When a kernel slides over an image, it can extend **beyond the edges**. Padding **adds extra pixels** (usually zeros) around the image to **preserve spatial information** at the borders.  

🔹 **Why is it important?**  
✅ Ensures important features at the edges are not lost.  
✅ Helps maintain **output size**, preventing a rapid reduction in feature map dimensions.  
✅ Enables **deeper architectures** by keeping feature maps larger for more layers.  

🔹 **Common Padding Techniques:**  
- **Zero Padding** → Adds zeros around the image (most common).  
- **Reflect Padding** → Mirrors the edge pixels.  
- **Replication Padding** → Repeats the edge pixels.  

📌 Example:  
- **Without Padding** → A 28×28 image becomes 26×26 after a **3×3** kernel is applied.  
- **With Padding (1 pixel)** → The image stays **28×28** after convolution.

---

## **2️⃣ Kernel Size: Feature Extraction Power**  
🔹 **What is it?**  
- The **kernel size (or filter size)** defines how much of the image the filter looks at **per step**.  
- Example sizes: **3×3, 5×5, 7×7**  

🔹 **Effect of Kernel Size:**  
✅ **Smaller kernels (e.g., 3×3)**  
   - Extract more **fine-grained details** (edges, textures).  
   - Allows stacking **more layers**, leading to better hierarchical feature learning.  

✅ **Larger kernels (e.g., 7×7, 9×9)**  
   - Capture **broader patterns** but **lose fine details**.  
   - Reduce feature map size **faster**, which can harm performance.  

📌 **Best Practice:**  
- **3×3 and 5×5 kernels** are commonly used in deep CNNs (e.g., VGG16, ResNet).  
- **Stacking multiple 3×3 layers** is better than using one large kernel (captures details efficiently).

---

## **3️⃣ Stride: Controlling Movement**  
🔹 **What is it?**  
- The **stride** defines how many pixels the kernel moves **after each step**.  
- Example:  
   - **Stride = 1** → Moves **1 pixel** at a time (fine feature extraction).  
   - **Stride = 2** → Moves **2 pixels** at a time (reduces spatial size faster).  

🔹 **Impact of Stride:**  
✅ **Small stride (e.g., 1)**  
   - Extracts **more features** since the kernel covers more overlapping regions.  
   - Generates **larger output feature maps**.  

✅ **Larger stride (e.g., 2 or more)**  
   - Reduces output **size faster** (useful for reducing computational cost).  
   - Extracts **fewer features**.  

📌 **Best Practice:**  
- Most modern CNNs **use stride = 1** for convolutions to capture **as much information as possible**.  
- Larger strides are often used in **pooling layers** to reduce dimensions efficiently.  

---

### **🔹 Summary: How to Tune These Hyperparameters?**
| Hyperparameter | Effect | Best Practice |
|---------------|--------|--------------|
| **Padding** | Preserves border information | Use **zero padding** to maintain size |
| **Kernel Size** | Defines the area the filter covers | Use **3×3 or 5×5** for deep networks |
| **Stride** | Controls how much the filter moves | Use **stride = 1** for fine details |

#create a convolutonal neural network

````python
import torch.nn as nn

class FashionMNISTModelV2(nn.Module):
    def __init__(self, input_shape: int, hidden_units: int, output_shape: int):
        super().__init__()
        # Create a conv layer
        self.conv_block1 = nn.Sequential(
            nn.Conv2d(
                in_channels=input_shape, out_channels=hidden_units,
                kernel_size=3, stride=1, padding=1  # Values we can set ourselves are called hyperparameters
            ),
            nn.ReLU(),
            nn.Conv2d(
                in_channels=hidden_units, out_channels=hidden_units,
                kernel_size=3, stride=1, padding=1
            ),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
        )

        self.conv_block2 = nn.Sequential(
            nn.Conv2d(
                in_channels=hidden_units, out_channels=hidden_units,
                kernel_size=3, stride=1, padding=1
            ),
            nn.ReLU(),
            nn.Conv2d(
                in_channels=hidden_units, out_channels=hidden_units,
                kernel_size=3, stride=1, padding=1
            ),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
        )

        self.classifier = nn.Sequential(
            nn.Flatten(),                      #10*7*7
            nn.Linear(                         # the output of conv block2 went into classifier model
                in_features=hidden_units*7*7,  # note below
                 out_features=output_shape
            )
        )

    def forward(self, x):
        x = self.conv_block1(x)
        #print(x.shape)
        x = self.conv_block2(x)
        #print(x.shape)#got 7*7 from here becuz inner matrices must match
        x = self.classifier(x)
        return x
````

We multiply `7*7` with `in_features` in the `nn.Linear` layer because the convolutional layers reduce the spatial dimensions (height and width), and we need to flatten the feature map before passing it to the fully connected (linear) layer.  

### **Understanding Why We Multiply `7*7`**  

1. **Initial Image Shape**:  
   - Your input image starts as **`(batch_size, channels, 28, 28)`** (e.g., **`(10, 1, 28, 28)`** for a batch of 10 grayscale images).  

2. **Convolution & Pooling Reduce Size**:  
   - **First Conv Block**:  
     - **Convolution**: Keeps the size the same (due to padding).  
     - **MaxPooling(2x2)**: Reduces size → **`(10, hidden_units, 14, 14)`**  
   - **Second Conv Block**:  
     - **Convolution**: Keeps the size the same.  
     - **MaxPooling(2x2)**: Reduces size → **`(10, hidden_units, 7, 7)`**  

3. **Flattening Before Fully Connected Layer**:  
   - After the last convolutional block, the tensor shape is **`(batch_size, hidden_units, 7, 7)`**.  
   - The `Flatten()` layer reshapes this to **`(batch_size, hidden_units * 7 * 7)`** (e.g., **`(10, hidden_units * 7 * 7)`**).  
   - This flattened vector is then fed into the `nn.Linear()` layer.  

4. **Why `hidden_units * 7 * 7` in `nn.Linear`?**  
   - `nn.Linear(in_features=hidden_units * 7 * 7, out_features=output_shape)` ensures that every value in the **flattened** feature map is connected to the next layer’s neurons.  
   - Without this multiplication, the fully connected layer wouldn't correctly map the extracted features to the output classes.  

### **Summary**  
We multiply `7*7` because after convolution and pooling, the spatial dimensions are reduced to `7x7`. Since we need a **1D vector** for the fully connected layer, we **flatten** the 3D feature map `(hidden_units, 7, 7)` into a single vector of size `hidden_units * 7 * 7`.

**Convolutional Neural Networks (CNNs) and Data Compression**

### **1. Purpose of CNNs in Deep Learning**
CNNs are designed to extract and **compress** input data into a **meaningful representation** that best suits the problem at hand. This compression helps in reducing computational cost, removing redundant information, and enhancing generalization.

### **2. Why Compression is Important?**
- **Reduces Computational Cost** – CNNs extract only essential features instead of processing the raw image.
- **Removes Redundancy** – Eliminates unnecessary background noise and irrelevant details.
- **Enhances Generalization** – The model learns key patterns that can apply to multiple inputs rather than memorizing specifics.

### **3. How CNNs Compress the Input?**
1. **Convolutional Layers** – Extract features from the image, starting with low-level details (edges, textures) and progressing to high-level representations (shapes, objects).
2. **Pooling Layers** – Reduce the spatial dimensions of feature maps while preserving essential information.
3. **Deeper Layers** – Combine extracted features into a more abstract, compact form.
4. **Fully Connected Layers** – Transform the final extracted features into a vector that represents the input effectively.

### **4. Example: Face Recognition**
- **Early layers** detect basic features like edges (nose, eyes, mouth).
- **Deeper layers** recognize complex facial patterns and structures.
- **Final layer** compresses this information into a unique vector representing the person's identity.

### **5. Key Takeaways**
- CNNs **compress** input data into a **structured, meaningful representation**.
- The network gradually **filters out unnecessary details**, focusing on patterns that help in classification or detection.
- **Efficient compression** enables CNNs to generalize well to new data while keeping computational costs manageable.

This process is fundamental to many applications, such as **image classification, object detection, and facial recognition**.

When you use `image.unsqueeze(0)`, it adds a batch dimension to the input. This works because PyTorch expects inputs to CNN models to be in the shape:

```
(batch_size, channels, height, width)
```

### **Understanding the Issue**
- `image.shape` was likely **`(1, 28, 28)`** (i.e., a single image without a batch dimension).
- Your model expects **`(batch_size, channels, height, width)`**.
- When you passed `image` directly, PyTorch tried to interpret `(1, 28, 28)` as `(batch_size=1, height=28, width=28)`, missing the `channels` dimension.
- This caused a **shape mismatch** in the convolutional layers.

### **Why `unsqueeze(0)` Works**
- `image.unsqueeze(0)` changes the shape from **`(1, 28, 28)`** to **`(1, 1, 28, 28)`**, correctly adding the missing batch dimension.
- Now, the shape matches what the model expects: **`(batch_size=1, channels=1, height=28, width=28)`**.

### **Alternative Fix**
Instead of using `.unsqueeze(0)`, ensure your dataset returns images with the correct shape (e.g., using `transforms.ToTensor()`, which automatically adds the channel dimension).

````python
def make_predictions(
    model:torch.nn.Module,
    data:list,
    device:torch.device):
  pred_probs=[]
  model.to(device)
  model.eval()
  with torch.inference_mode():
    for sample in data:
      #Prepare the sample (add a batch dimention and pass to target device)
      sample = torch.unsqueeze(sample,dim=0).to(device)

      #forward pass (model output raw logits)
      pred_logit=model(sample)

      #get prediction (logit-> prediction probabilities)
      pred_prob=torch.softmax(pred_logit.squeeze(),dim=0)

      #store prediction probabilities
      pred_probs.append(pred_prob.cpu())
  return torch.stack(pred_probs)






````

#making a confusion matrix for evaluation



A **confusion matrix** is a table that helps evaluate the performance of a classification model by comparing the predicted labels with the actual labels. It provides insights into **how well the model is classifying each class** and helps identify **misclassifications**.

---

### **Structure of a Confusion Matrix**
For a classification task with two classes (**binary classification**), the confusion matrix looks like this:

| Actual \ Predicted | Predicted **Positive** | Predicted **Negative** |
|--------------------|----------------------|----------------------|
| **Actual Positive**  | **True Positive (TP)**  | **False Negative (FN)** |
| **Actual Negative**  | **False Positive (FP)** | **True Negative (TN)** |

For a **multi-class classification** problem (e.g., FashionMNIST with 10 classes), the confusion matrix is a square matrix where each row represents the **actual** class and each column represents the **predicted** class.

---

### **Key Terms**
- **True Positive (TP)** → Model correctly predicts a positive class.  
- **True Negative (TN)** → Model correctly predicts a negative class.  
- **False Positive (FP)** → Model incorrectly predicts a positive class when it's actually negative (Type I error).  
- **False Negative (FN)** → Model incorrectly predicts a negative class when it's actually positive (Type II error).  

---

### **How to Interpret?**
1. **Perfect Model** → All values are along the diagonal (high **TP** and **TN**, zero **FP** and **FN**).
2. **High FN** → Model is **missing** actual positives (bad recall).
3. **High FP** → Model is **misclassifying** negatives as positives (bad precision).

---

### **Performance Metrics Derived from the Confusion Matrix**
1. **Accuracy** = \( \frac{TP + TN}{TP + TN + FP + FN} \)  
2. **Precision** = \( \frac{TP}{TP + FP} \) (How many of the predicted positives were actually correct?)  
3. **Recall (Sensitivity)** = \( \frac{TP}{TP + FN} \) (How many of the actual positives were correctly identified?)  
4. **F1 Score** = \( 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}} \) (Harmonic mean of Precision & Recall)

---
